```python
# Separate data based on the 'Class' column for analysis
legit = credit_card_data[credit_card_data['Class'] == 0]
fraud = credit_card_data[credit_card_data['Class'] == 1]
```

```python
print("Shape of legitimate transactions:", legit.shape)
print("Shape of fraudulent transactions:", fraud.shape)
```

```python
# statistical details 
legit['Amount'].describe()
```

```python
fraud['Amount'].describe()
```

```python
## compare the values of both transactions
credit_card_data.groupby('Class').mean()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot([legit['Amount'], fraud['Amount']], labels=['Legit', 'Fraud'])
plt.title('Boxplot of Transaction Amounts')
plt.ylabel('Amount')
plt.show()
```

```python
credit_card_data['Class'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Class Distribution: Legit (0) vs Fraud (1) ')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(credit_card_data['Amount'], bins=50, color='purple')
plt.title('Transaction Amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

## correlation
Logistic regression assumes that independent variables are not highly correlated.
So checking correlation helps you detect and fix multicollinearity early

```python
credit_card_data.corr()["Class"].sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,5))

sns.kdeplot(legit["Time"], label="Legit", fill=True)
sns.kdeplot(fraud["Time"], label="Fraud", fill=True)

plt.xlabel("Time (seconds since first transaction)")
plt.title("Density Plot of Time for Legit vs Fraud")
plt.legend()
plt.show()
```

```python
credit_card_data["Time_hours"] = credit_card_data["Time"] / 3600
```

```python
bins = [0, 6, 12, 18, 24]
labels = ["0-6 hrs", "6-12 hrs", "12-18 hrs", "18-24 hrs"]

credit_card_data["Time_bucket"] = pd.cut(credit_card_data["Time_hours"], bins=bins, labels=labels)
```

```python
time_bucket_counts = credit_card_data.groupby(["Time_bucket", "Class"]).size().unstack()
print(time_bucket_counts)
```

```python
time_bucket_counts.plot(kind="bar", figsize=(8,5))

plt.xlabel("Time Buckets")
plt.ylabel("Number of Transactions")
plt.title("Fraud vs Legit Transactions Across Time Buckets")
plt.show()
```

Insight:
The time-based analysis shows that fraudulent transactions are not uniformly distributed over time. Certain time windows exhibit a higher concentration of fraud transactions compared to others, indicating potential behavioral patterns that could be useful for modeling.
